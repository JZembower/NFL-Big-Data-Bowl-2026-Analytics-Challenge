# 2026 NFL Big Data Bowl - Data Dictionary

## Overview
This document provides a comprehensive summary of the datasets in the 2026 Big Data Bowl. The tracking data is provided by the NFL Next Gen Stats team.

---

## External Data Policy
The 2026 Big Data Bowl allows participants to use external NFL data as long as it is **free and publicly available** to all participants. 

**Approved Sources:**
- [nflverse](https://nflverse.nflverse.com/)
- [Pro Football Reference](https://www.pro-football-reference.com/)

> **Note:** The `game_id` and `play_id` of the Big Data Bowl data merges with the `old_game_id` and `play_id` of nflverse's play-by-play data.

---

## Dataset Files

### 1. Input Data: `input_2023_w[01-18].csv`

**Description:** Contains tracking data **before** the pass is thrown.

**Key Variables:**

| Variable | Type | Description |
|----------|------|-------------|
| `game_id` | Numeric | Game identifier, unique |
| `play_id` | Numeric | Play identifier, not unique across games |
| `player_to_predict` | Boolean | Whether the x/y prediction for this player will be included in the output file |
| `nfl_id` | Numeric | Player identification number, unique across players |
| `frame_id` | Numeric | Frame identifier for each play/type, starting at 1 for each game_id/play_id/file type |
| `play_direction` | Text | Direction that the offense is moving (left or right) |
| `absolute_yardline_number` | Numeric | Distance from end zone for possession team |
| `player_name` | Text | Player name |
| `player_height` | Text | Player height (ft-in) |
| `player_weight` | Numeric | Player weight (lbs) |
| `player_birth_date` | Date | Birth date (yyyy-mm-dd) |
| `player_position` | Text | The player's position (specific role on field) |
| `player_side` | Text | Team player is on (Offense or Defense) |
| `player_role` | Text | Role player has on play (Defensive Coverage, Targeted Receiver, Passer, or Other Route Runner) |
| `x` | Numeric | Player position along the long axis of the field (0-120 yards) |
| `y` | Numeric | Player position along the short axis of the field (0-53.3 yards) |
| `s` | Numeric | Speed in yards/second |
| `a` | Numeric | Acceleration in yards/second² |
| `o` | Numeric | Orientation of player (degrees) |
| `dir` | Numeric | Angle of player motion (degrees) |
| `num_frames_output` | Numeric | Number of frames to predict in output data for the given game_id/play_id/nfl_id |
| `ball_land_x` | Numeric | Ball landing position along the long axis of the field (0-120 yards) |
| `ball_land_y` | Numeric | Ball landing position along the short axis of the field (0-53.3 yards) |

---

### 2. Output Data: `output_2023_w[01-18].csv`

**Description:** Contains tracking data **after** the pass is thrown.

**Key Variables:**

| Variable | Type | Description |
|----------|------|-------------|
| `game_id` | Numeric | Game identifier, unique |
| `play_id` | Numeric | Play identifier, not unique across games |
| `nfl_id` | Numeric | Player identification number, unique across players |
| `frame_id` | Numeric | Frame identifier for each play/type, starting at 1. Max value matches `num_frames_output` from input file |
| `x` | Numeric | Player position along the long axis of the field (0-120 yards) |
| `y` | Numeric | Player position along the short axis of the field (0-53.3 yards) |

---

### 3. Supplementary Data: `supplementary_data.csv`

**Description:** Contextual information about the game and play.

#### Game Information

| Variable | Type | Description |
|----------|------|-------------|
| `game_id` | Numeric | Game identifier, unique |
| `season` | Numeric | Season of game |
| `week` | Numeric | Week of game |
| `game_date` | Date | Game date (mm/dd/yyyy) |
| `game_time_eastern` | Time | Start time of game (HH:MM:SS, EST) |
| `home_team_abbr` | Text | Home team three-letter code |
| `visitor_team_abbr` | Text | Visiting team three-letter code |
| `home_final_score` | Numeric | Total points scored by home team |
| `visitor_final_score` | Numeric | Total points scored by visiting team |

#### Play Information

| Variable | Type | Description |
|----------|------|-------------|
| `play_id` | Numeric | Play identifier, not unique across games |
| `play_description` | Text | Description of play |
| `quarter` | Numeric | Game quarter |
| `game_clock` | Time | Time on clock of play (MM:SS) |
| `down` | Numeric | Down |
| `yards_to_go` | Numeric | Distance needed for a first down |
| `possession_team` | Text | Team abbr of team on offense with possession |
| `defensive_team` | Text | Team abbr of team on defense |
| `yardline_side` | Text | 3-letter team code corresponding to line-of-scrimmage |
| `yardline_number` | Numeric | Yard line at line-of-scrimmage |
| `pre_snap_home_score` | Numeric | Home score prior to the play |
| `pre_snap_visitor_score` | Numeric | Visiting team score prior to the play |

#### Pass Information

| Variable | Type | Description |
|----------|------|-------------|
| `pass_result` | Text | Dropback outcome (C: Complete, I: Incomplete, S: Sack, IN: Intercepted, R: Scramble) |
| `play_nullified_by_penalty` | Text | Whether penalty cancels play outcome (Y/N) |
| `pass_length` | Numeric | Distance beyond LOS that ball traveled (negative if behind LOS) |
| `offense_formation` | Text | Formation used by possession team |
| `receiver_alignment` | Text | Enumerated as 0x0, 1x0, 1x1, 2x0, 2x1, 2x2, 3x0, 3x1, 3x2 |
| `route_of_targeted_receiver` | Text | Route ran by targeted receiver |
| `play_action` | Binary | Indicates if there was play action on the play |

#### Quarterback Information

| Variable | Type | Description |
|----------|------|-------------|
| `dropback_type` | Text | Type of drop back (Traditional, Designed Rollout, Scramble, etc.) |
| `dropback_distance` | Numeric | Distance QB dropped back (yards) behind center after snap |
| `pass_location_type` | Text | Location of QB at time of throw (Inside Tackle Box, Outside Left, Outside Right, Unknown) |

#### Defense Information

| Variable | Type | Description |
|----------|------|-------------|
| `defenders_in_the_box` | Numeric | Number of defenders in close proximity to line-of-scrimmage |
| `team_coverage_man_zone` | Text | Overarching type of coverage (Man/Zone) |
| `team_coverage_type` | Text | Specific kind of coverage assigned on the play |

#### Outcome Metrics

| Variable | Type | Description |
|----------|------|-------------|
| `penalty_yards` | Numeric | Yards gained by offense by penalty |
| `pre_penalty_yards_gained` | Numeric | Net yards gained by offense, before penalty yardage |
| `yards_gained` | Numeric | Net yards gained by offense, including penalty yardage |
| `expected_points` | Numeric | Expected points on this play |
| `expected_points_added` | Numeric | Delta of expected points on this play |
| `pre_snap_home_team_win_probability` | Numeric | Win probability of home team before the play |
| `pre_snap_visitor_team_win_probability` | Numeric | Win probability of visiting team before the play |
| `home_team_win_probability_added` | Numeric | Win probability delta for home team |
| `visitor_team_win_probility_added` | Numeric | Win probability delta for visitor team |

---

## Key Join Variables

To merge datasets, use the following key variables:

- **Primary Keys:** `game_id` + `play_id`
- **Player Tracking:** `game_id` + `play_id` + `nfl_id` + `frame_id`
- **External Data:** `game_id` ↔ `old_game_id` (nflverse), `play_id` ↔ `play_id` (nflverse)

```python
# Load the datasets of choice, can change the weeks input
input_df = pd.read_csv('/Users/jonahzembower/Downloads/NFL Big Data Bowl 2026 Analytics Challenge/Kaggle Data/train/input_2023_w01.csv')
output_df = pd.read_csv('/Users/jonahzembower/Downloads/NFL Big Data Bowl 2026 Analytics Challenge/Kaggle Data/train/output_2023_w01.csv')
supp_df = pd.read_csv('/Users/jonahzembower/Downloads/NFL Big Data Bowl 2026 Analytics Challenge/Kaggle Data/supplementary_data.csv')

print("="*80)
print("DATASET SHAPES AND BASIC INFO")
print("="*80)
print(f"\nInput Data Shape: {input_df.shape}")
print(f"Output Data Shape: {output_df.shape}")
print(f"Supplementary Data Shape: {supp_df.shape}")

print("\n" + "="*80)
print("MISSING VALUES ANALYSIS")
print("="*80)

# Create missing values summary
def missing_summary(df, name):
    missing = pd.DataFrame({
        'Column': df.columns,
        'Missing_Count': df.isnull().sum().values,
        'Missing_Percentage': (df.isnull().sum().values / len(df) * 100).round(2)
    })
    missing = missing[missing['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
    if len(missing) == 0:
        missing = pd.DataFrame({'Column': ['No missing values'], 'Missing_Count': [0], 'Missing_Percentage': [0.0]})
    return missing

input_missing = missing_summary(input_df, "Input")
output_missing = missing_summary(output_df, "Output")
supp_missing = missing_summary(supp_df, "Supplementary")

print("\n--- INPUT DATA MISSING VALUES ---")
print(input_missing.to_string(index=False))

print("\n--- OUTPUT DATA MISSING VALUES ---")
print(output_missing.to_string(index=False))

print("\n--- SUPPLEMENTARY DATA MISSING VALUES ---")
print(supp_missing.to_string(index=False))

print("\n" + "="*80)
print("KEY STATISTICS")
print("="*80)

# Unique plays and players
print(f"\nInput Data:")
print(f"  - Unique games: {input_df['game_id'].nunique()}")
print(f"  - Unique plays: {input_df['play_id'].nunique()}")
print(f"  - Unique game-play combinations: {input_df.groupby(['game_id', 'play_id']).ngroups}")
print(f"  - Unique players: {input_df['nfl_id'].nunique()}")
print(f"  - Players to predict: {input_df['player_to_predict'].sum()}")
print(f"  - Total frames: {len(input_df)}")
print(f"  - Avg frames per player per play: {input_df.groupby(['game_id', 'play_id', 'nfl_id']).size().mean():.2f}")

print(f"\nOutput Data:")
print(f"  - Unique games: {output_df['game_id'].nunique()}")
print(f"  - Unique plays: {output_df['play_id'].nunique()}")
print(f"  - Unique game-play combinations: {output_df.groupby(['game_id', 'play_id']).ngroups}")
print(f"  - Unique players: {output_df['nfl_id'].nunique()}")
print(f"  - Total frames: {len(output_df)}")
print(f"  - Avg frames per player: {output_df.groupby(['game_id', 'play_id', 'nfl_id']).size().mean():.2f}")

print(f"\nSupplementary Data:")
print(f"  - Unique games: {supp_df['game_id'].nunique()}")
print(f"  - Unique plays: {supp_df['play_id'].nunique()}")
print(f"  - Unique game-play combinations: {len(supp_df)}")

# Player roles and positions
print("\n" + "="*80)
print("PLAYER DISTRIBUTION")
print("="*80)
print("\nPlayer Roles:")
print(input_df.groupby('player_role')['nfl_id'].nunique().sort_values(ascending=False))

print("\nPlayer Positions (Top 10):")
print(input_df.groupby('player_position')['nfl_id'].nunique().sort_values(ascending=False).head(10))

print("\nPlayer Side:")
print(input_df.groupby('player_side')['nfl_id'].nunique())

# Pass results
print("\n" + "="*80)
print("PLAY OUTCOMES")
print("="*80)
print("\nPass Results:")
print(supp_df['pass_result'].value_counts())

print("\nCoverage Types:")
print(supp_df['team_coverage_type'].value_counts().head(10))
```

```python
# Fix the outlier detection for supplementary data
print("="*80)
print("OUTLIER ANALYSIS - SUPPLEMENTARY DATA (Key Metrics)")
print("="*80)

supp_numeric = ['yards_gained', 'expected_points', 'expected_points_added', 
                'pass_length', 'dropback_distance', 'defenders_in_the_box']

def detect_outliers_series(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((series < lower_bound) | (series > upper_bound)).sum()
    return outliers, lower_bound, upper_bound

supp_outlier_summary = []
for col in supp_numeric:
    if col in supp_df.columns and supp_df[col].notna().sum() > 0:
        series = supp_df[col].dropna()
        outliers, lower, upper = detect_outliers_series(series)
        supp_outlier_summary.append({
            'Column': col,
            'Outliers': outliers,
            'Outlier_Pct': f"{(outliers/len(series)*100):.2f}%",
            'Lower_Bound': f"{lower:.2f}",
            'Upper_Bound': f"{upper:.2f}",
            'Min': f"{series.min():.2f}",
            'Max': f"{series.max():.2f}"
        })

supp_outlier_df = pd.DataFrame(supp_outlier_summary)
print("\n", supp_outlier_df.to_string(index=False))

print("\n" + "="*80)
print("CORRELATION ANALYSIS - INPUT DATA")
print("="*80)

# Correlation for input data
input_corr = input_df[['x', 'y', 's', 'a', 'dir', 'o', 'frame_id', 'ball_land_x', 'ball_land_y']].corr()
print("\nTop Correlations (absolute value > 0.3):")
corr_pairs = []
for i in range(len(input_corr.columns)):
    for j in range(i+1, len(input_corr.columns)):
        if abs(input_corr.iloc[i, j]) > 0.3:
            corr_pairs.append({
                'Variable 1': input_corr.columns[i],
                'Variable 2': input_corr.columns[j],
                'Correlation': f"{input_corr.iloc[i, j]:.3f}"
            })

if corr_pairs:
    corr_df = pd.DataFrame(corr_pairs).sort_values('Correlation', key=lambda x: abs(x.astype(float)), ascending=False)
    print(corr_df.to_string(index=False))
else:
    print("No strong correlations found (|r| > 0.3)")

print("\n" + "="*80)
print("CORRELATION ANALYSIS - SUPPLEMENTARY DATA")
print("="*80)

supp_corr_cols = ['yards_gained', 'expected_points', 'expected_points_added', 
                  'pass_length', 'dropback_distance', 'defenders_in_the_box',
                  'pre_snap_home_score', 'pre_snap_visitor_score']
supp_corr = supp_df[supp_corr_cols].corr()

print("\nTop Correlations (absolute value > 0.3):")
supp_corr_pairs = []
for i in range(len(supp_corr.columns)):
    for j in range(i+1, len(supp_corr.columns)):
        if abs(supp_corr.iloc[i, j]) > 0.3:
            supp_corr_pairs.append({
                'Variable 1': supp_corr.columns[i],
                'Variable 2': supp_corr.columns[j],
                'Correlation': f"{supp_corr.iloc[i, j]:.3f}"
            })

if supp_corr_pairs:
    supp_corr_df = pd.DataFrame(supp_corr_pairs).sort_values('Correlation', key=lambda x: abs(x.astype(float)), ascending=False)
    print(supp_corr_df.to_string(index=False))
```

### 3. MOVEMENT PATTERNS
- ✓ Route runners & targeted receivers move fastest (avg 4.0 yards/sec)  
- ✓ Passers slowest (avg 1.8 yards/sec) - mostly stationary in pocket  
- ✓ Route runners have highest acceleration (avg 2.6 yards/sec²)  
- ✓ Speed increases with frame_id (r=0.57) - players accelerate during play  
- ✓ Very few outliers in speed/acceleration (<1%) - clean tracking data  

### 4. SPATIAL PATTERNS
- ✓ Strong correlation between player X and ball landing X (r=0.875)  
- ✓ Players concentrate in middle third of field (162k of 285k frames)  
- ✓ Ball lands in endzone 11,501 times (4% of frames)  
- ✓ Ball thrown behind LOS 137k times - likely screen passes or QB under pressure  
- ✓ Average ball travel: only 0.51 yards (suggests many short/screen passes)  

### 5. PLAY CONTEXT
- ✓ Pass completions: 69.3% (12,470 of 18,009 plays)  
- ✓ Incompletions: 28.3% (5,106 plays)  
- ✓ Interceptions: 2.4% (433 plays)  
- ✓ Cover 3 Zone most common (31.4%), followed by Cover 1 Man (22.8%)  
- ✓ Cover 3 has highest completion rate, Cover 1 Man more balanced

- ✓ Acceleration: 2,037 outliers (0.71%) - sudden direction changes  
- ✓ Yards gained: 677 outliers (3.76%) - big plays and sacks  
- ✓ EPA: 729 outliers (4.05%) - high-leverage situations  
- ✓ Pass length: 978 outliers (5.43%) - deep shots downfield  
- ✓ Most outliers are legitimate extreme plays, not data errors  

### 7. KEY CORRELATIONS
- ✓ Yards gained ↔ EPA: r=0.734 (strong positive)  
- ✓ Player X ↔ Ball landing X: r=0.875 (very strong)  
- ✓ Speed ↔ Frame ID: r=0.572 (moderate positive)  
- ✓ Home/Visitor scores: r=0.404 (competitive games)  

### 8. MODELING RECOMMENDATIONS

**→ Feature engineering:**
  - Distance to ball landing position  
  - Velocity vectors (combine speed + direction)  
  - Relative positions to other players  
  - Coverage type indicators  
  - Time-series features (frame-to-frame changes)  

**→ Model considerations:**
  - Sequence models (LSTM/GRU) for temporal dependencies  
  - Separate models for offense vs defense (different movement patterns)  
  - Consider player role as key feature (4x speed difference)  
  - Ball landing position is strong predictor of player movement  

**→ Data splits:**
  - Split by game_id to avoid data leakage  
  - Consider stratifying by pass_result for balanced evaluation  
  - Week 1 only - be cautious of early season patterns  

### 9. POTENTIAL CHALLENGES
- ⚠ Ball travel distance anomaly: avg 0.51 yards suggests data issue or coordinate system confusion (absolute vs relative yardline)  
- ⚠ Supplementary data has 17,190 plays not in input/output (historical context only - can be used for feature engineering)  
- ⚠ Direction/orientation in degrees (0-360) may need circular encoding  
- ⚠ Field coordinates may need normalization for play_direction
